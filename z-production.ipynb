```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# constants
a = 1/137.
mZ = 91.1876 # GeV
GammaZ = 2.4952 # GeV
GF = 1.166 * 10**-5 # /GeV^2

# These are wrong for now
cVe = 1
cAe = 1
cVmu = 1
cAmu = 1
```

```python
# Chi(s)

def ReX(s, GammaZ=GammaZ):
    return GF*mZ*mZ/(8*np.pi*np.sqrt(2))*s*(s-mZ*mZ)/((s-mZ*mZ)*(s-mZ*mZ) + s*s*GammaZ*GammaZ/(mZ*mZ))

def ImX(s, GammaZ=GammaZ):
    return GF*mZ*mZ/(8*np.pi*np.sqrt(2))*s*(-1*s*GammaZ/mZ)/((s-mZ*mZ)*(s-mZ*mZ) + s*s*GammaZ*GammaZ/(mZ*mZ))
```

```python
x = np.linspace(0,300,101)
```

```python
plt.plot(x,ReX(x*x), label='Real part')
plt.plot(x,ImX(x*x), label='Imaginary part')
plt.plot(x,ReX(x*x)*ReX(x*x) + ImX(x*x)*ImX(x*x), label='Norm squared')
plt.xlabel('sqrt(s)')
plt.ylabel('chi(s)')
plt.suptitle('Real Z width')
plt.legend(frameon=False)
```

```python
plt.plot(x,ReX(x*x,0),label='Real part')
plt.plot(x,ImX(x*x,0),label='Imaginary part')
plt.plot(x,ReX(x*x,0)*ReX(x*x,0) + ImX(x*x,0)*ImX(x*x,0),label='Norm squared')

plt.xlabel('sqrt(s)')
plt.ylabel('chi(s)')
plt.suptitle('Zero Z width')
plt.legend(frameon=False)
```

```python
def sigma_gamma(s):
    return np.pi*a*a/(2*s)*(3*np.pi/2)

def sigma_Z(s, GammaZ=GammaZ):
    return 8*np.pi/(s) * (ReX(s,GammaZ)*ReX(s,GammaZ) + ImX(s,GammaZ)*ImX(s,GammaZ))*((cVe*cVe + cAe*cAe)*(cVmu*cVmu + cAmu*cAmu)*3*np.pi/2)
                                
def sigma_interference(s,GammaZ=GammaZ):
    return -4*np.pi*a/(s) * ReX(s,GammaZ) *(cVe*cVmu*3*np.pi/2)
```

```python
x = np.linspace(0,300,101)
```

```python
plt.plot(x,sigma_gamma(x*x),label='$\sigma_\gamma$')
plt.plot(x,sigma_Z(x*x), label='$\sigma_Z$')
plt.plot(x,sigma_interference(x*x),label='Interference')

plt.plot(x,sigma_gamma(x*x) + sigma_interference(x*x) + sigma_Z(x*x),label='Total')

plt.xlabel('sqrt(s)')
plt.ylabel('Cross section')
plt.suptitle('Real Z width')
plt.legend(frameon=False)
```

```python
plt.plot(x,sigma_gamma(x*x),label='$\sigma_\gamma$')
plt.plot(x,sigma_Z(x*x,0),label='$\sigma_Z$')
plt.plot(x,sigma_interference(x*x,0),label='Interference')

plt.plot(x,sigma_gamma(x*x) + sigma_interference(x*x,0) + sigma_Z(x*x,0),label='Total')

plt.xlabel('sqrt(s)')
plt.ylabel('Cross section')
plt.suptitle('Zero Z width')
plt.legend(frameon=False)
```
